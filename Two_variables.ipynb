# Two variable plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.0, 100.0, 1001)
y = x

plt.plot(x,y)
```

```python
x = np.linspace(0.0, 100.0, 1001)
y1 = 2.0 * x + 5.0
y2 = 1.0 * x - 5.0
y3 = -2.0 * x + 0.0

plt.plot(x, y1, 'r')
plt.plot(x, y2, 'b')
plt.plot(x, y3, 'g')
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
x = np.linspace(-100.0, 100.0, 1001)
y1 = x * x - 4.0 * x + 4.0

plt.plot(x, y1, 'b')
```
